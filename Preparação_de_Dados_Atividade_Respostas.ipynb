```python
import pandas as pd
import seaborn as sns

# Carregamento  do dataset
df = sns.load_dataset('diamonds')
df.head()
```

### 1.1 - Preencha os valores nulos de depth com a mediana. Coloque o resultado em um novo dataframe.

```python
df_adjusted = pd.DataFrame(df) # ou df_adjusted = df.copy()
df_adjusted['depth'] = df_adjusted['depth'].fillna(df_adjusted['depth'].median())
df_adjusted.head()
```

### 1.2 - Remova linhas com valores nulos. Use o dataframe anterior e atualize-o.

```python
df_adjusted = df_adjusted.dropna()
df_adjusted.head()
```

### 1.3 - Remova linhas duplicadas. Use o dataframe anterior e atualize-o.

```python
df_adjusted = df_adjusted.drop_duplicates()
df_adjusted.head()
```

### 1.4 - Normalize a coluna price usando Min-Max (nova coluna). Use o dataframe anterior e atualize-o.

```python
def min_max(serie) -> pd.Series:
    return (serie - serie.min()) / (serie.max() - serie.min())

df_adjusted['price_normalized'] = min_max(df_adjusted['price'])
df_adjusted.head()
```

### 1.5 - Normalize a coluna carat usando z-score (nova coluna). Use o dataframe anterior e atualize-o.

```python
def z_score(serie) -> pd.Series:
    return (serie - serie.mean()) / serie.std()

df_adjusted['carat_normalized'] = z_score(df_adjusted['carat'])
df_adjusted.head()
```

### 1.6 - Remova outliers de carat usando IQR. Use o dataframe anterior e CRIE um novo dataframe.

```python
Q1 = df_adjusted['carat'].quantile(0.25)
Q3 = df_adjusted['carat'].quantile(0.75)
IQR = Q3 - Q1
df_outlier = df_adjusted[(df_adjusted['carat'] >= Q1 - 1.5 * IQR) & (df_adjusted['carat'] <= Q3 + 1.5 * IQR)]
df_outlier
```

### 1.7 - Remova outliers de price com Z-Score (> 3). Use o dataframe anterior e atualize-o.

```python
import scipy as sp
import numpy as np

z_scores = sp.stats.zscore(df_outlier['price'])

df_outlier = df_outlier[np.abs(z_scores) <= 3]
df_outlier
```

### 1.8 - Crie uma coluna caro onde True se o preço for maior que 5000. Use o dataframe anterior.

```python
df_outlier['caro'] = df_outlier['price'] > 5000
df_outlier
```

### 1.9 - Agrupe por cut e color e agregue-os por média (mean), menor (min) e maior (max) para price e carat. Use o dataframe anterior.

```python
df_outlier.groupby(['cut', 'color'], observed = True)[['price', 'carat']].agg(['mean', 'min', 'max'])
```

### 1.10 - Quantos diamantes possuem mais de 1 quilate (carat > 1.0) e corte "Premium". Use o dataframe anterior.

```python
db_quilates = df_outlier.query('carat > 1.0 and cut == "Premium"')
len(db_quilates)
```
