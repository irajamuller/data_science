Iremos utilizar o dataset <strong>Titanic</strong> do Seaborn. O objetivo é visualizar os dados do dataset Titanic.
</p>


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Carregar o dataset Titanic
titanic = sns.load_dataset("titanic")
```

### 1.1 - Crie um gráfico de barras com a contagem de passageiros por classe (eixo X).

```python
df_classe = titanic['class'].value_counts().sort_index()

plt.figure(figsize=(6, 4))
plt.bar(df_classe.index, df_classe.values, color=["blue", 'white', 'black'], edgecolor="black")

# Configura opções gerais de vizualização do gráfico
plt.title("Número de Passageiros x Classe Social", fontsize=16)
plt.xlabel("Classe Social", fontsize=12)
plt.ylabel("Número de Passageiros", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xticks(['First', 'Second', 'Third'], ['1ª classe', '2ª classe', '3ª classe'])
plt.show()
```

### 1.2 - Crie um gráfico de linhas que mostre quantos sobreviventes havia por idade (eixo X).

```python
df_sobreviventes = titanic[titanic['survived'] == True]
df_sobreviventes = df_sobreviventes['age'].value_counts().sort_index()

plt.figure(figsize=(6, 4))
plt.plot(df_sobreviventes.index, df_sobreviventes.values)

plt.title("Sobreviventes x Idade", fontsize=16)
plt.xlabel("Idade", fontsize=12)
plt.ylabel("Número de Sobreviventes", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### 1.3 - Crie um gráfico de caixas da idade por sexo (eixo X). Remova tuplas não disponíveis somente para as séries/colunas idade e sexo.

```python
df_idade_sexo = titanic.dropna(subset=['age', 'sex'])

plt.figure(figsize=(6, 4))

sns.boxplot(df_idade_sexo, x='sex', y='age')

plt.title("Idade x Gênero", fontsize=16)
plt.xlabel("Gênero", fontsize=12)
plt.ylabel("Idade", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

### 1.4 - Crie um gráfico de dispersão entre idade (eixo X) e tarifa, colorindo por sobrevivência. Remova tuplas não disponíveis somente para as séries/colunas idade, tarifa e sobreviveu.

```python
df_idade_tarifa = titanic.dropna(subset=['age', 'fare', 'survived'])

plt.figure(figsize=(6, 4))

# Coloquei cores para mostrar quem sobreviveu
colors = df_idade_tarifa['survived'].map({0: 'red', 1: 'green'})
plt.scatter(df_idade_tarifa['age'], df_idade_tarifa['fare'], c=colors, alpha=0.7, edgecolor="black")

plt.title("Idade x Tarifa", fontsize=16)
plt.xlabel("Idade", fontsize=12)
plt.ylabel("Tarifa", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.show()
```

### 1.5 - Crie uma matriz heatmap para todas as séries que são numéricas. Deverá ser usada no dataframe a função corr() que é o cálculo de correlação (coeficiente de correlação de Pearson).

```python
plt.figure(figsize=(6, 4))
df_correlacao = titanic[['pclass', 'age', 'sibsp', 'parch', 'fare']]

sns.heatmap(df_correlacao.corr(), annot=True, cmap="Blues", fmt=".2f", cbar=True, square=True)

plt.title("Matriz Confusão - Colunas Numéricas", fontsize=14)
plt.show()
```
